```python
from typing import List
import pandas as pd
import json

def to_jsonl(data: List[dict], filename: str):
    with open(filename, "w") as f:
        for item in data:
            f.write(json.dumps(item) + "\n")

def to_txt(data: str, filename: str):
    with open(filename, "w") as f:
        f.write(data)
```

```python
bank_schema = pd.read_excel("resources/database/schemas/bank.xlsx", sheet_name="bank")
bank_schema = bank_schema.fillna("")
bank_schema.head()
```

```python
for table in bank_schema["Table Name"].unique():
    to_txt(
        bank_schema[bank_schema["Table Name"] == table].to_markdown(index=False),
        f"resources/database/schemas/{table}.txt"
    )
to_txt(
    bank_schema.to_markdown(index=False),
    "resources/database/schemas/bank.txt"
)
```

```python
solution_schema = pd.read_excel("resources/database/schemas/bank.xlsx", sheet_name="solution")
solution_schema = solution_schema.fillna("")
solution_schema["Table Name"].unique()
for table in solution_schema["Table Name"].unique():
    to_txt(
        solution_schema[solution_schema["Table Name"] == table].to_markdown(index=False),
        f"resources/database/schemas/{table}.txt"
    )
to_txt(
    solution_schema.to_markdown(index=False),
    "resources/database/schemas/solution.txt"
) 
```
